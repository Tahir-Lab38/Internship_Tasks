# Task 1: Exploring and Visualizing the Iris Dataset

## ðŸŽ¯ Objective
To load, inspect, and visualize the Iris dataset to understand the structure, relationships, and distributions of its features. This helps in developing an intuition about the dataset before model building.


```python
# Importing Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')

# Basic Inspection
print("âœ… Shape of dataset:", df.shape)
print("âœ… Column names:", df.columns.tolist())

# Display first few rows
df.head()
```

```python
# Info and Summary Stats
print("ðŸ“‹ Dataset Info:")
df.info()

print("\nðŸ“Š Descriptive Statistics:")
df.describe()
```

```python
# Scatter plot to show relationships between features
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', palette='Set1')
plt.title('Scatter Plot: Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid(True)
plt.show()
```

```python
# Histograms of all numeric features
df.hist(figsize=(10, 8), edgecolor='black', grid=False)
plt.suptitle("Histograms of All Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Box plots for all features
plt.figure(figsize=(12, 6))
for i, column in enumerate(df.columns[:-1]):  # exclude 'species'
    plt.subplot(1, 4, i+1)
    sns.boxplot(y=df[column], color='lightblue')
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()
```

## ðŸ” Insights and Conclusion

- **Petal length and petal width** show strong clustering based on species.
- **Sepal measurements** overlap more between species, especially between versicolor and virginica.
- Some mild outliers may be present in sepal width.
- Distributions of features differ in skewness and spread, especially petal features.

This exploratory analysis will serve as a foundation for applying ML models later.
